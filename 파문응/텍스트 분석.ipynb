### 전체 데이터프레임에 적용시키려면 apply 함수 사용

```python
import string #특수문자를 전처리하기 위해 string 임포트

#전처리 하는 함수 
def str_punc(x):
    pre_str=[]
    for i in x:
        if i not in string.punctuation:
            pre_str.append(i)
    pre_str = ''.join(pre_str)
    return pre_str
```

```python
#apply 함수를 전체에 적용
df['text'] = df['text'].apply(str_punc)
```

### 문자열 데이터를 수치화로 만들면서 시각적인 인사이트 비교

```python
#토큰에 대한 길이
tk_sp_len = [len(t) for t in tk_sp]
```

```python
#전체 문장에 대한 길이로 수치를 만들어 보자!
tk_t_sp_len = [len(s.replace(' ',"")) for s in df['text']]
```

```python
# 토큰화와 전체 문장의 길이를 비교하며 시각적으로 두 개를 비교분석해본 그래프입니다!

plt.figure(figsize=(15,10))
plt.hist(tk_sp_len, bins=50, alpha=0.5, color='r', label='token')
plt.hist(tk_t_sp_len, bins=50, alpha=0.5, color='b', label='all')
plt.yscale('log')
plt.savefig('tklen12.png') #plt.savefig('파일명')을 통해 시각화 이미지로 저장 가능
```

## 정규식을 통해 전처리 진행

```python
sp_list= list(data.text)
```

```python
sp_list
```

```python
# 정규화 진행하면 -> 정규식으로도 간단하게 전처리가 가능하다.
# 내가 원하는 전처리 -> 디테일을한 부분을 잡으려면 정규식을 많이 사용한다.
import re
list_reg = []
for i in data.text:
    text = re.sub('[^a-zA-Z0-9]',' ',i).strip()
    if (text !=''):
        list_reg.append(text)
```

```python
list_reg
```

```python
#1차 정규식으로 전처리한 데이터를 불용어를 지워보자!
res_sp=[ word for word in list_reg if word not in eg_stop ] 
```

```python
res_sp
```

```python
nltk.download('punkt')
```

```python
#토크나이저를 이용해서 나눠보자!
token_sp= word_tokenize(str(res_sp))
```

```python
token_sp
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
#명사다 동사나, 형용사 이런 것들을 비교해 볼 수 있다.
my_tag_set = ['NN','VB'] #그외에 다른 품사 태깅을 넣어서 추출을 할 수 있다.
my_tk_sp = [word for word, tag in nltk.pos_tag(token_sp) if tag in my_tag_set ]
```

```python
my_tk_sp
```

```python
# 위의 문장들을 실제 카운팅을 진행해 보자!
sp_cnt = dict()
for word in my_tk_sp:
    sp_cnt[word] = sp_cnt.get(word, 0)+1
```

```python
sp_cnt
```

- 해당 토큰과 숫자들을 정리하였으니 -> 간단한 워드클라우드로 시각화 진행

```python
wc=WordCloud(width= 500, height = 500, scale =2.0, max_font_size=200, background_color='gray')
```

```python
wc  #워드클라우드로 데이터가 보여지는 형식은 카운팅 형식
```

```python
wc.generate_from_frequencies(sp_cnt)
plt.figure()
plt.imshow(wc)
```

- 데이터 분석의 하나의 과정속에서 진행하는 형태
- 워드 클라우드가 어떤 분석에 대한 결과를 보여주는 것보다는 분석에 진행하는 과정 중에 하나의 소스
- 고객데이터, 리뷰 데이터 등을 가지고 전처리 하면서 어떤 특징들이 있는지 시각화로 보여줄 수 있다.
