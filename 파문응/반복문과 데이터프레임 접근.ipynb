# 반복문 복습

- 반복문 : 데이터프레임 전처리 과정에서 다양한 데이터를 빠르게 전처리하기 위해서는 반복문이 필요하다!

```python
for i in range(5):
    print(i)
```

```python
a=['국','준','호','국준','국호','준호','국준호']
b=['빨','주','노','초','파','남','보','무지개']
```

```python
for i in a:
    print(i)
```

```python
for i in range(len(a)):
    print(a[i])    # 인덱스로 가져오기
```

```python
for i,j in enumerate(a):
    print(i,j)      # enumerate 함수 - 앞 : 인덱스 값, 뒤 : 실제 값 으로 출력
```

```python
for i,j in zip(a,b):
    print(i,j)    #i->a, j->b에 대한 루프인덱스
```

```python
for i ,(j,k) in enumerate(zip(a,b)):
    print(i,j,k)   #루프인덱스 + enumerate 응용 가능
```

```python
import itertools
#라이브러리 중 중첩반복문을 사용하지 않고도 모든 경우를 출력할 수 있다.
for i,j in itertools.product(a,b):
    print(i,j)
```

## 반복문을 데이터프레임에 적용해보자

```python
import pandas as pd

df = pd.read_csv('mpg.csv')
```

```python
df
```

```python
for i in df:
    print(i)
```

```python
# 단순히 데이터프레임의 컬럼 값들을 출력하는 것은 의미 X
# 반복문을 응용하자
```

```python
# 이 두개의 연비 평균을 내면, 차량의 평균 연비가 나온다.
for i,j in zip(df.cty, df.hwy):
    print((i+j)/2)
```

```python
# 데이터프레임에 새로운 칼럼을 넣어보자
mc = [] # -->연비 평균 칼럼 추가하자
for i,j in zip(df.cty, df.hwy):
    mc.append((i+j)/2)
```

```python
df['mc']=mc  #데이터프레임에 새로운 칼럼 넣는 법
```

```python
df
```

```python
df.info()
#데이터 분석시 문자열, 수치형에 대해서는 분석전에 전처리 또는 시각화하는 방법이 다르다.
#데이터 타입에 따라서 데이터를 나눠서 보는 것도 좋은 방법 중 하나다!
```

```python
df['year'].dtype #각 칼럼별 데이터타입도 출력할 수 있다.
```

```python
for i in df.columns:
    print(df[i].dtype)
```

```python
for i in df.columns:
    df_type = df[i].dtype
    if df_type =='object':
        print(i, df_type)
        
#타입에 대한 것을 변수에 넣고, 칼럼과 데이터타입 함께 출력
```

```python
for i in df.columns:
    df_type = df[i].dtype
    if df_type =='object':
        print(i, df[i].value_counts())
        print(i, df[i])
        
#문자형데이터는 카운팅을 해보자! 카운팅의 값을 보여줘!
```
