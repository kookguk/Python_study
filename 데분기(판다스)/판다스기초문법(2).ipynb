```python
## 성별이 남성인 경우 여성인 경우에 따라서 반복문과 if문을 이용해서 0, 1 매핑하기!
## 값을 지정해서 들어가면 [], loc, iloc으로 값을 찾았다.
## 여성과 남성인지를 확인하기 위해서
##tt.sex.loc[3]
##남성 0 여성이면 1

##이런 새로운 변수를 만들 경우 파생변수를 하나 만들어서 진행할 수 있다.

tt['sex_onehot']= 0
for i in range(len(tt)):
    if tt.sex.loc[i] == 'male':
        tt['sex_onehot'].loc[i] = 0
    elif tt.sex.loc[i]== 'female':
        tt['sex_onehot'].loc[i] = 1
```

```python
print(tt.sex_onehot.value_counts())
```

```python
print(tt.sex.value_counts())
```

## group by

- 집계하는 것 
- 기준 컬럼을 가지고 통계치를 보고 싶은 컬럼의 값을 볼 수 있다.
- Groupby 하나의 기준 컬럼의 그룹으로 만든다.

```python
#pclass 1,2,3 등급으로 나뉘어져 있음
grouped=tt.groupby('class')
```

```python
grouped
```

```python
for i, j in grouped:
    print(i)
    print(j.head(10))
```

- get_group(내가원하는 그룹)

```python
grouped.get_group('First')
```

- groupby 그룹을 두 개 이상 진행하게 되면
- []

```python
groups=tt.groupby(['class','embarked'])
```

```python
groups
```

```python
for i, j in groups:
    print(i)
    print(j.head(10))
```

```python
groups.mean()
```

```python
groups.std()
```

- groupby agg 함수를 넣어서 통계치를 살펴본다.

- groupby 그룹을 묶고 내가 원하는 통계치들을 정리해서 볼 수 있다.

- tt.groupby('그룹할컬럼1').agg(변수명 = ('기준컬럼1','기준컬럼을 볼 통계치'), 추가추가..)

```python
tt
```

```python
tt.groupby('class').agg(fare_sum = ('fare','sum'))
```

```python
tt.groupby('class').agg(fare_sum = ('fare','sum'),
                       fare_mean= ('fare','mean'),
                       fare_median = ('fare','median'),
                       n_cnt= ('fare','count'))
```

- 나만의 튜닝으로 만든 함수를 agg에 넣어서 값을 계산하자!


```python
def min_max(x):
    return x.max()- x.min()
```

```python
grouped.agg(min_max)
```

```python
# 1개의 컬럼의 2개 이상 통게치를 볼 때
grouped.agg({'fare':['sum','median','mean','std','var']})
```

## 데이터 병합

- concat, merge 
- concat : 덩어리와 덩어리가 합쳐진다. df1, df2 합쳐진다. 날 것 그대로 
    - 행으로 붙이는 것 df1 + df2
    - 열로 붙이는 것 
    - df1
    - '+'
    - df2
- merge : join 형태, left, outer, right join 개념

```python
# 데이터프레임으로 만들어서 두 문법의 원리를 이해해보자!
test1= pd.DataFrame({'학회원번호':[1,2,3,4,5],
                    '학회원분반':['Python','SQL','R','데이터분석입분','데이터분석전처리']})
```

```python
test2= pd.DataFrame({'학회원번호':[1,2,3,4,5],
                    '벌점현황':[5,5,10,0,3]})
```

```python
#pd.merge(왼쪽컬럼(기준), 오른쪽컬럼(그다음), how='병합방법', on='공통기준컬럼')
# left_on, right_on 지정할 수 있음
pd.merge(test1, test2, how='inner', on='학회원번호')
```

```python
test1
```

```python
# pd.concat을 이용해서 데이터 병합
# pd.concat(axis = 1,0 에 따라 달라진다. 1 오른쪽으로 붙여진다. 0 컬럼별로 붙여진다.)
pd.concat([test1, test2])
```

```python
pd.concat([test1, test2,test1,test2], axis=1) # 컬럼이 옆으로 쌓여서 붙여진다.
```

```python
pd.concat([test1, test2, test1], axis=0) #컬럼 기준 위아래 붙여진다. # 행 열 
```
