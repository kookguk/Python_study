```python
df

# 내가 만들 그래프를 위해 컬럼을 추출하자

# 나는 countplot으로 그림을 그릴래 - > 카운트가 어떻게 되는지?

interested = ['sex','embarked','class','alive']

fig, axs =plt.subplots(nrows=2, ncols=2, figsize=(12,6))

for idx, variable in enumerate(interested):
    ax = axs[idx//2, idx%2]
    sns.countplot(x=variable, hue='who', data=df, ax=ax)
    
plt.show()
```

```python
##  리스트와 함수들을 딕셔너리 매핑해서 가능하다.


fig, axs =plt.subplots(nrows=2, ncols=2, figsize=(12,6))

plots = [
    {'func' : sns.barplot, 'x':'sex', 'y':'survived', 'title':'Survived Rate by sex', 'ax' : axs[0,0]},
    {'func' : sns.barplot, 'x':'class', 'y':'survived', 'title':'Survived Rate Rate by class', 'ax' : axs[0,1]},
    {'func' : sns.histplot, 'x':'age',  'title':'Survived Rate by age', 'ax' : axs[1,0]},
    {'func' : sns.histplot, 'x':'fare',  'title':'Survived Rate by fare', 'ax' : axs[1,1]}]


for plot in plots:
    plot['func'](data=df, x=plot.get('x'), y=plot.get('y',None), ax=plot['ax'])
    plot['ax'].set_title(plot['title'])
    
## 그래프 충돌방지
plt.tight_layout() 
plt.show()
```
